```python
# Coletando os dados a serem utilizados

# Precisei ler o arquivo dessa forma
df = pd.read_csv('dados.csv', sep=";", encoding='ISO-8859-1')
```

```python
# Utilizando o group by para avaliar
# Esse groupby eu juntei por produtos que foram comprados iguais, portanto juntei por clientes que compraram os mesmos produtos.

df.groupby(['PRODUTOS']).size()
```

```python
# Criando uma lista vazia
lista = []

# Loop For para passar por todas linhas do DataFrame
# Dentro do try vamos colocar tudo o que queremos que aconteça e se por acaso se der algum tipo de erro ele vai cair no except e vai passar.
# Depois do try, temos a primeira linha Print que é o que vamos colocar na tela e embaixo dela o que vamos jogar dentro da lista. O teste é
# o nome de uma variável qualquer que dei o nome, o i é por onde meu código vai passar, por exemplo vai ver o nome de cada palavra e o SPLIT ele 
# vai separar as STRINGS, através da vírgula, sempre que ele encontrar uma vírgula ele vai separar. Entao no primeiro PRODUTO, vai ler abacate, 
# logo depois de abacate, através do split ele vai ler uma vírgula e vai separar como uma string. 

# O que vai ser armazenado dentro da variável teste vai ser uma lista e vamos armazenar dentro de LISTA, tudo o que está dentro do w.

for i in df['PRODUTOS']:
    try:
        print(i.split(','))
        teste = i.split(',')
        for w in teste:
            lista.append(w)
    except:
        pass  
```

```python
# Verificando a lista

lista
```

```python
# Converto minha lista em um DataFrame

df_lista = pd.DataFrame(lista, columns = ['PRODUTOS'])
df_lista.head(5)
```

```python
df_lista.shape
```

```python
# Fazer um GroupBy para contar os produtos e ordenar por ordem decrescente

df_lista.groupby(['PRODUTOS']).size().sort_values(ascending = False)
```

### Agora precisamos criar nossas variáveis pra cada produto

Para cada item do nosso dataframe precisa virar uma variável para o nosso algoritmo de machine learning que vamos criar. Cada produto será uma variável.

```python
# Criando um dicionario de dados com os produto
# Deixamos igual mas poderiamos caso se os nomes originais fossem inglês, poderíamos fazer a tradução. 

dicionario = {'DETERGENTE':'DETERGENTE',               
            'AMACIANTE':'AMACIANTE',       
            'CHOCOLATE':'CHOCOLATE',       
            'CEBOLA':'CEBOLA',          
            'ARROZ':'ARROZ',           
            'FEIJAO':'FEIJAO',           
            'CAFÉ':'CAFÉ',             
            'MELAO':'MELAO',            
            'VINAGRE':'VINAGRE',          
            'ACUCAR':'ACUCAR',           
            'REFRIGERANTE':'REFRIGERANTE',     
            'FRALDAS':'FRALDAS',          
            'MANDIOCA':'MANDIOCA',         
            'SUCO':'SUCO',             
            'BATATA':'BATATA',           
            'TOMATE':'TOMATE',           
            'CANDIDA':'CANDIDA',          
            'ALHO':'ALHO',             
            'OVOS':'OVOS',             
            'PERA':'PERA',             
            'UVA':'UVA',              
            'MELANCIA':'MELANCIA',         
            'ABACATE':'ABACATE' } 

# Visualizando o dicionário

dicionario
```

```python
# Visualizando nosso DataFrame

df.head(10)
```

```python
# Gerando os campos no DataFrame
# ITERROWS vai retornar a linha e o indice do meu dataframe, no caso primeiro será o index que vou colocar em index, e segundo
# va iretornar linha que vou colocar em row.

for index, row in df.iterrows():    
    for chave, valor in dicionario.items():
        try:            
            if chave in row['PRODUTOS']:
                df.at[index, valor] = 'Sim'
            else:
                df.at[index, valor] = 'Nao'
        except:
            pass
```
